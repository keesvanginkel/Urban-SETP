### Prepare the sea level rise scenarios from the source data

#### Le Bars, D., Drijfhout, S., & de Vries, H. (2017). A high-end sea level rise probabilistic projection including rapid Antarctic ice sheet mass loss. Environmental Research Letters, 12(4), 044013. https://doi.org/10.1088/1748-9326/aa6512

Properties of this dataset:
 - Global sea level rise projection
 - Sea level rise relative to 1995

```python
import os as os
import pandas as pd
```

```python
source_path = os.path.join("LeBars_2017.xlsx")
df = pd.read_excel(source_path,header=1,index_col=0,usecols="A:G")
df = df * 10**-2 #convert units to [m]
df
```

```python
colnames = ['45_05','45_med','45_95','85_05','85_med','85_95']
df.columns = ['LeBars_2017_RCP' + x for x in colnames]
```

```python
for column in df.columns:
    dest_path = os.path.join(column + '.csv')
    df[column].to_csv(dest_path,header=False)
```
